## Range

```python
x=range(10)
```

```python
x
```

```python
for i in x:
    print(i)
```

```python
x=range(5,20)
for i in x:
    print (i)
```

```python
x=range(5,20, 3) # 3rd case
for i in x:
    print (i)
```

```python
x=range(20,5,-3) # 4th case reverse
for i in x:
    print (i)
```
